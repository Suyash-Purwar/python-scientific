```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")
```

```python
# Loading in data from csv
data = pd.read_csv("../matplotlib/data4.csv")
```

```python
# Filtering and storing data 
views = data["view_count"]
likes = data["likes"]
ratio = data["ratio"]
```

```python
# Plotting scatter plots
plt.scatter(views, likes, c=ratio, cmap="afmhot", edgecolor="black", linewidth=1)
colorbar = plt.colorbar()
colorbar.set_label("Like/Dislike Ratio")
plt.rcParams["figure.figsize"] = [7, 5]

# Assigning plotting scales
plt.xscale("log")
plt.yscale("log")

# Assigning labels on plot
plt.title("Correlation of views and likes")
plt.xlabel("Views")
plt.ylabel("Likes")

plt.tight_layout()

plt.show()
```
